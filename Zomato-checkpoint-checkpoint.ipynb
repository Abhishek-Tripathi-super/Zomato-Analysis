```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load Dataset
df = pd.read_csv("Indian-Resturants.csv")
```

```python
# Display first few rows
display(df.head())
```

```python
# Basic Info
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Handling Missing Values
df = df.dropna()
```

```python
# Descriptive Statistics
display(df.describe())
```

```python
# Distribution of Restaurant Ratings
plt.figure(figsize=(8, 5))
sns.histplot(df, x='aggregate_rating', bins=20, kde=True, hue='price_range', palette='coolwarm')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python
# City-wise Restaurant Count
plt.figure(figsize=(12, 6))
top_cities = df['city'].value_counts().head(10).index
df_top_cities = df[df['city'].isin(top_cities)]

sns.countplot(x='city', hue='price_range', data=df_top_cities, palette=('coolwarm'))
plt.title('Top 10 Cities with Most Restaurants (Grouped by Price Range)')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.legend(title='Price Range')
plt.show()
```

```python
# Cuisine Analysis
plt.figure(figsize=(10, 5))
cuisine_counts = df['cuisines'].value_counts().head(10)
sns.barplot(y=cuisine_counts.index, x=cuisine_counts.values, hue=cuisine_counts.index, palette='viridis')
plt.title('Top 10 Most Popular Cuisines')
plt.xlabel('Count')
plt.ylabel('Cuisines')
plt.show()
```

```python
# Price Range vs. Ratings
plt.figure(figsize=(8,6))
sns.boxplot(x=df['price_range'], y=df['aggregate_rating'], palette='muted')
plt.title("Price Range vs Ratings")
plt.xlabel("Price Range")
plt.ylabel("Rating")
plt.show()
```

```python
# Online Order vs Ratings
plt.figure(figsize=(6, 4))
sns.countplot(x='delivery', hue='price_range', data=df, palette='coolwarm')
plt.title('Impact of Online Orders on Ratings')
plt.xlabel('Online Order Available')
plt.ylabel('Count')
plt.show()
```

```python
# Table Booking Analysis
plt.figure(figsize=(6, 4))
sns.countplot(x='opentable_support', hue='price_range', data=df, palette='magma')
plt.title('Table Booking vs Ratings')
plt.xlabel('Table Booking Available')
plt.ylabel('Count')
plt.show()
```

```python
# Top Restaurant Chains
plt.figure(figsize=(10, 5))
top_chains = df['name'].value_counts().head(10).index
df_top_chains = df[df['name'].isin(top_chains)]
sns.countplot(y='name', hue='price_range', data=df_top_chains, palette='coolwarm')
plt.title('Top 10 Restaurant Chains')
plt.xlabel('Number of Outlets')
plt.ylabel('Restaurant Name')
plt.show()
```

```python
# Word Cloud
text = " ".join(review for review in df['rating_text'].dropna())
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Customer Review Word Cloud")
plt.show()
```

### **Summary of the Analysis:**

1. **Data Overview & Cleaning:**
   - The dataset was loaded and inspected for missing values.
   - Missing values were removed to ensure clean data for analysis.
   - Basic statistics were examined to understand distributions.

2. **Restaurant Ratings Distribution:**
   - A histogram was used to visualize the **distribution of restaurant ratings**.
   - Most restaurants have ratings clustered around **4.0 to 4.5**, indicating generally favorable reviews.

3. **City-wise Restaurant Count:**
   - The **top 10 cities** with the most restaurants were analyzed.
   - Cities like **Delhi, Mumbai, and Bangalore** have the highest concentration of restaurants.
   - Price ranges vary significantly across cities.

4. **Cuisine Analysis:**
   - The **top 10 most popular cuisines** were identified.
   - Indian cuisine dominates, but **continental and Chinese cuisines** are also quite popular.

5. **Price Range vs. Ratings (Top 50 Restaurants):**
   - A **boxplot** was used to visualize how **price range** influences restaurant ratings.
   - Higher price range restaurants tend to have **better ratings** on average.

6. **Impact of Online Orders on Ratings:**
   - Restaurants that offer **online orders** have a diverse range of ratings.
   - The presence of online ordering does not always guarantee higher ratings.

7. **Table Booking Analysis:**
   - The **availability of table booking** was analyzed.
   - The number of restaurants that provide table booking is relatively lower compared to those that do not.

8. **Top Restaurant Chains:**
   - The **top 10 most popular restaurant chains** were identified.
   - Well-known brands dominate in terms of the number of outlets.

9. **Word Cloud for Reviews:**
   - A **word cloud** was generated from customer reviews.
   - Common words include **"great", "delicious", "service", "friendly"**, indicating positive sentiments.

### **Key Takeaways:**
- Cities with a higher number of restaurants tend to have a **diverse price range and cuisines**.
- **Higher-end restaurants** (higher price range) generally receive **better ratings**.
- The presence of **table booking and online ordering** does not necessarily lead to **higher ratings**.
- **Indian and fast-food cuisines** dominate the market.
- **Top restaurant chains** have a strong presence and brand recognition.
- Customer reviews indicate that **food quality and service** play a crucial role in restaurant success.
